```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('./airbnb_hw.csv', low_memory=False)
print( df.shape, '\n')
df.head()
```

```python
price = df['Price']
price.unique()
```

```python
price = df['Price']
price = price.str.replace(',','') # Replace commas with nothing
price = pd.to_numeric(price,errors='coerce') # Typecast price to float/numeric
print( price.unique() , '\n')
print( 'Total missing: ', sum( price.isnull() ) )
#after conversion to float no values are lost even those >999

#after observing the series of values replacing commas with spaces and converting to a float makes the price variable clean and avoids missing values
```

```python
df = pd.read_csv('./sharks.csv', low_memory=False)
df.head()
df.columns.tolist()
# based off of the list there are 233 unamed columns indicating this is a very messy dataset
```

```python
df['Type'].value_counts()
```

```python
type = df['Type']

type = type.replace(['Sea Disaster', 'Boat','Boating','Boatomg'],'Watercraft') # All watercraft/boating values
type.value_counts()

type = type.replace(['Invalid', 'Questionable','Unconfirmed','Unverified','Under investigation'],'Uncertain') # consolidates unclean values into one, I could have dropped these but found they were similar and significant enough to keep
type.value_counts()

df['Type'] = type # Replace the 'Type' variable with the cleaned version
del type # Destroy the temporary vector

df['Type'].value_counts()
```

```python
df = pd.read_csv('./pretrial2017.csv', low_memory=False)
df.head()
```

```python
df['WhetherDefendantWasReleasedPretrial'].value_counts() #there are 31 missing values to be dropped
```

```python
release = df['WhetherDefendantWasReleasedPretrial']
release = release.replace(9,np.nan) # replacing all unavailable values
df['WhetherDefendantWasReleasedPretrial'] = release # updating the variable
del release
```

```python
df['ImposedSentenceAllChargeInContactEvent'].value_counts()
```

```python
df['SentenceTypeAllChargesAtConvictionInContactEvent'].value_counts()
# 8779 cases with charges dismissed
# 274 unavailable values
# these two values could account for the 9053 '' in the imposed sentence variable
```

```python
length = df['ImposedSentenceAllChargeInContactEvent']
type = df['SentenceTypeAllChargesAtConvictionInContactEvent']

length = length.mask( type == 4, 0) # where the value is 4 in sentence type make the imposed yrs value = 0
length = length.mask( type == 9, np.nan) # where the value is 9 in sentence type remove the value in imposed

length_NA = length.isnull() # Create a new missing dummy
print( pd.crosstab(length_NA, type), '\n')
print( np.sum(length_NA),'\n') # 274 missing, much better



df['ImposedSentenceAllChargeInContactEvent'] = length # Replace data with cleaned version
del length, type # Delete temporary length/type variables
```
